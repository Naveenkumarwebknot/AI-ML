```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('./tripadvisor_hotel_reviews.csv')
df.head()
```

**Transform the 'Rating' column to nominal data type with binary labels**

```python
Rating_binary_label=pd.get_dummies(df["Rating"])
Rating_binary_label
```

**Name labels 'positive' for rating value > 3 and 'negative' for <= 3**

```python
dr=df.copy()
dr['rating_changed']=df['Rating']
dr['rating_changed']=dr['rating_changed'].apply(lambda x:'positive' if x>3 else 'negative')
dr
```

```python
df
```

**Calculate the lengths of each record in 'review' column and add to the dataframe**

```python
dr=dr.copy()
dr['review_length']=dr['Review'].apply(lambda x:len(x))
dr
```

```python
dr=dr.copy()
dr['Word_Count'] = dr['Review'].apply(lambda x: len(x.split()))
dr
```

Calculate the average length of word for each record in 'review' column and add to the dataframe

```python
dr=dr.copy()

dr['Avg_Word_Length'] = dr['Review'].apply(lambda x: sum(len(word) for word in x.split()) / len(x.split()))
dr
```

 **Calculate the average sentence length of the sentences in the 'review' and add to the dataframe**

```python
dr=dr.copy()
dr['Avg_sentence_len']=dr['Review'].apply(lambda x:len(x.split()))/dr['Review'].apply(lambda x:len(x.split(".")))
dr
```

**Visualise each of the columns with respect to the 'Rating' column**

```python

import seaborn as sns
import matplotlib.pyplot as plt


numeric_columns = ['review_length', 'Word_Count', 'Avg_Word_Length','Avg_sentence_len']


for col in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Rating', y=col, data=dr)
    plt.title(f'{col} vs Rating')
    plt.show()
```

**Compute term frequency analysis on the 'review' column,**

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk import FreqDist
import matplotlib.pyplot as plt

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
df1=df.copy()

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return tokens
df1['Review_lists'] = df1['Review'].apply(preprocess_text)
all_words = [word for sublist in df1['Review_lists'] for word in sublist]
freq_dist = FreqDist(all_words)
most_common_words = freq_dist.most_common(10)

plt.figure(figsize=(10, 6))
freq_dist.plot(30, cumulative=False)

plt.show()

```
