```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./tested.csv')
```

```python
df.head()
```

**Perform one-hot encoding: replace the categorical variable by different boolean variables (0/1) to indicate whether or not certain label is true for that observation**

```python
df1=df.copy()
df_encoded = pd.get_dummies(df, columns=['Sex','Pclass'])
df_encoded
```

**Perform ordinal - encoding: replace the labels by some ordinal number if ordinal is meaningful**

```python
df2=df.copy()

categorical_columns = ['Sex', 'Pclass']


for col in categorical_columns:
    df2[col + '_ordinal_encoded'] = df2[col].astype('category').cat.codes

df2
```

**Perform Mean encoding: replace the label by the mean of the target for that label (the target must be 0/1 valued or continuous)**

```python
df3=df.copy()
categorical_columns = ['Sex', 'Pclass']


for col in categorical_columns:
    mean_encoded = df3.groupby(col)['Survived'].mean()
    df3[col + '_mean_encoded'] = df3[col].map(mean_encoded)
df3
```

**Perform target-encoding: similar to mean encoding, but use both posterior probability and probability of the target**

```python
 pip install category_encoders
```

```python
import category_encoders as ce
df4=df.copy()
categorical_columns = ['Sex', 'Pclass']


encoder = ce.TargetEncoder()


for col in categorical_columns:
    df4[col + '_target_encoded'] = encoder.fit_transform(df4[col], df4['Survived'])
df4
```

**Perform Weight of Evidence (WOE) encoding: replace the label with WOE of each label. WOE is computed from the basics odds ratio [ In((Proportion of Good Outcomes)/ (Proportion of Bad Outcomes) ]**


```python
df4=df.copy()
categorical_columns = ['Sex', 'Pclass']


for col in categorical_columns:

    proportion_good_outcomes = df4.groupby(col)['Survived'].mean()
    proportion_bad_outcomes = 1 - proportion_good_outcomes

    woe = np.log(proportion_good_outcomes / proportion_bad_outcomes)


    df4[col + '_woe_encoded'] = df4[col].map(woe)
df4
```
