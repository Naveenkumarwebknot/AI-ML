```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./tested.csv")
```

```python
df.head()
```

**Get the dtypes for each column and general data description of the entire dataset:**

```python
dtype_col={}
for col in df:
  dtype_col[col]=df[col].dtypes
dtype_col
```

```python
description=df.describe()
description
```

**Draw the barplot of a discrete variable x=pclass against target variable y=survived. What is the default value shown wrt to y?**

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

Draw the countplot of a discrete variable x=pclass

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Count of Passengers by Passenger Class')
plt.show()
```

**Draw the boxplot of a discrete variable x=pclass against y=fare**

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass',y='Fare',data=df)
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

**Draw the distplot of a continuous variable x=fare**

```python
plt.figure(figsize=(8, 6))
sns.distplot(df['Fare'])
plt.title('Distribution of Fare')
plt.show()
```

**Draw the scatter plot of two variables, x=fare and y=pclass**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Fare', y='Pclass', data=df)
plt.title('Scatter Plot of Fare vs Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Plot')
plt.show()
```
