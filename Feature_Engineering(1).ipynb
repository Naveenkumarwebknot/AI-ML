**Exclude all cases (listwise) with missing values**

```python
df_droped = df.dropna()
df_droped
```

**Create an additional column â€˜NAâ€™ indicating whether the data was missing for that observation**

```python
df1=df.copy(False)
df1['NA'] = df1.isnull().any(axis=1).astype(int)
df1
```

**Arbitrary Value Imputation - replace the NA by arbitrary values**

```python
df1.fillna(99)
df1
```

**Mean/Median/Mode Imputation - replace the NA be mean/median/mode of that column**

```python

df1_mean=df1.mean()
df1_filled_mean = df1.fillna(df1_mean)

df1
```

End of distribution Imputation - replace the NA by values that are at the far end of the distribution of that column (tip - consider it as Normal Distribution -> 3*std away from the mean) **bold text**

```python
df2=df.copy()
end_of_distribution=df2.mean()+3*df2.std()
df2.fillna(end_of_distribution)
```

**Random Imputation - Replace the NA with random sampling from the pool of available observation of the column**

```python
df3=df.copy()
for col in df3.columns:
    missing = df3[col].isnull()
    num_missing = missing.sum()
    if num_missing > 0:
        values = df3.loc[~missing, col]
        rand_index = np.random.choice(values.index, size=num_missing)
        df3.loc[missing, col] = values[rand_index].values
df3
```

**Identify outliers by IQR - Interquartile Range Rule**

```python
numeric_cols = df2.select_dtypes(include=np.number)
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)
outliers
```

**Winsorization - Top coding and bottom coding (capping the maximum of a distribution at an arbitrary set value, vice versa)**

```python

from scipy.stats.mstats import winsorize
df5=df.copy()
numeric_cols = df5.select_dtypes(include=np.number)


winsorization_percentage = 0.05


df5_winsorized = numeric_cols.apply(lambda x: winsorize(x, limits=(winsorization_percentage, winsorization_percentage)))


df5_winsorized = pd.DataFrame(df5_winsorized, columns=df.columns)

df5_winsorized

```
